```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_excel('Aviation_KPIs_Dataset.xlsx')
df.head()
```

```python
df.dropna()
```

```python
features = df.drop(columns=['Profit (USD)','Flight Number','Scheduled Departure Time','Actual Departure Time'])
target=df['Profit (USD)']
```

```python


from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=0.2, random_state=2023)
```

```python
X_val
```

```python
features
```

```python
from sklearn.preprocessing import StandardScaler,scale
scaler = StandardScaler()


#Fit the StandardScaler to the training data
scaler.fit(X_train)

# transform both the training and validation data
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)

```

```python
X_train
```

```python
hyper_params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': ['l1','l2'],
    'learning_rate': 0.005,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.7,
    'bagging_freq': 10,
    'verbose': 0,
    "max_depth": 8,
    "num_leaves": 128,  
    "max_bin": 512,
    "num_iterations": 100000
}
```

```python
from lightgbm import LGBMRegressor as lgb

gbm = lgb(**hyper_params)

```

```python
import pickle
with open('flight_profit.pkl', 'rb') as file:
    gbm = pickle.load(file)
```

```python

y_pred = gbm.predict(X_val)

df = pd.DataFrame(y_pred)
df.to_csv('predictions.csv')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(Y_val, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_val, y_pred)
print(f"R² Score: {r2}")

```

```python
import numpy as np

mape = np.mean(np.abs((Y_val - y_pred) / Y_val)) * 100
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y_val, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
import numpy as np

rmse = np.sqrt(mean_squared_error(Y_val, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
